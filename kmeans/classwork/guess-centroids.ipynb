```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.loadtxt('./points.csv', delimiter=",", skiprows=0)
```

```python
points.shape
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(points[:, 0], points[:, 1], color='dodgerblue', marker='o', edgecolors='k', label="Points", alpha=0.75)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True, linestyle="--", alpha=0.25)
plt.title('Points')
```

```python
x_min, x_max = points[:, 0].min(), points[:, 0].max()
y_min, y_max = points[:, 1].min(), points[:, 1].max()

print(f'X: min => {x_min:.4f}, max => {x_max:.4f}')
print(f'Y: min => {y_min:.4f}, max => {y_max:.4f}')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
plt.scatter(points[:, 0], points[:, 1], color='dodgerblue', marker='o', edgecolors='k', label="Points", alpha=0.75)
ax.plot([x_min, x_max, x_max, x_min, x_min],
        [y_min, y_min, y_max, y_max, y_min],
        color="purple", linestyle="--", label="Bounding Box", linewidth=1.5, alpha=0.75
        )
ax.set_title('Points XY min max')
ax.legend()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True, linestyle="--", alpha=0.25)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
plt.scatter(points[:, 0], points[:, 1], color='dodgerblue', marker='o', edgecolors='k', label="Points", alpha=0.75)
ax.plot([x_min, x_max, x_max, x_min, x_min],
        [y_min, y_min, y_max, y_max, y_min],
        color="purple", linestyle="--", label="Bounding Box", linewidth=1.5, alpha=0.75
        )
ax.scatter(-3, -3, marker='^', color='red', edgecolors='k', s=150, alpha=1)
ax.scatter(-2, 4, marker='^', color='red', edgecolors='k', s=150, alpha=1)
ax.scatter(0, 0, marker='^', color='red', edgecolors='k', s=150, alpha=1)
ax.scatter(5.15, 5.15, marker='^', color='red', edgecolors='k', s=150, alpha=1)
ax.set_title('Points with guessed centroids')
ax.legend()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True, linestyle="--", alpha=0.25)
```
