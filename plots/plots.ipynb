```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 40)

# Define functions
def f1(x):
    return x ** 2

def f2(x):
    return x * np.sin(2 * x)

def f3(x):
    return np.arctan(x)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, f1(x), label="f(x) = x", color="blue", linestyle="-", marker="o", markersize=4)
plt.plot(x, f2(x), label="f(x) = x * sin(2x)", color="green", linestyle="--", marker="s", markersize=4)
plt.plot(x, f3(x), label="f(x) = arctan(x)", color="red", linestyle=":", marker="^", markersize=4)
plt.xlabel("x", fontsize=13, fontweight='bold')
plt.ylabel("f(x)", fontsize=13, fontweight='bold')
plt.title("Function Plots", fontsize=13, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12, loc='best')

plt.ylim(-10, 50)

plt.tight_layout()
plt.show()
```
