```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

# Part 1

```python
# generate evenly spaced values within start and stop
# starts at -10, ends at 10
# generates 40 evenly spaced values
x = np.linspace(-10, 10, 40)
```

```python
# Define functions
def f1(x):
    return x ** 2

def f2(x):
    return x * np.sin(2 * x)

def f3(x):
    return np.arctan(x)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, f1(x), label="f(x) = x²", color="blue", linestyle="-", marker="o", markersize=4)
plt.plot(x, f2(x), label="f(x) = x * sin(2x)", color="red", linestyle="--", marker="s", markersize=4)
plt.plot(x, f3(x), label="f(x) = arctan(x)", color="purple", linestyle=":", marker="^", markersize=4)
plt.xlabel("x", fontsize=13, fontweight='bold')
plt.ylabel("f(x)", fontsize=13, fontweight='bold')
plt.title("Function Plots", fontsize=13, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12, loc='best')

plt.ylim(-10, 50)

plt.tight_layout()
plt.show()
```

```python
# load data from a csv to numpy array
points = np.loadtxt('./data/points.csv', delimiter=',', skiprows=0)
distances = np.loadtxt('./data/distances.csv', delimiter=',', skiprows=0)

print(f'Shape points: {points.shape}')
print(f'Shape distances: {distances.shape}')
```

```python
x = points[:, 0]
y = points[:, 1]

plt.figure(figsize=(10, 8))
# c = color the points based on distance values
# edgecolors = black border
scatter = plt.scatter(x, y, c=distances, cmap='viridis', edgecolors='k', alpha=0.75)

# color bar
plt.colorbar(scatter, label="Distance Value")

plt.xlabel('X points')
plt.ylabel('Y points')
plt.title('Scatter plot with color distances')
plt.grid(True)
plt.show()
```

# Part 2

```python
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)

print(f'Shape x: {x.shape}')
print(f'Shape y: {y.shape}')
```

```python
# creates 2D grid for numpy broadcasting
# X: contains copies of x values along rows
# Y: contains copies of y values along cols
X, Y = np.meshgrid(x, y)

print(f'Shape X: {X.shape}')
print(f'Shape Y: {Y.shape}')
```

```python
# define functions
def f1(x, y):
    return x * y

def f2(x, y):
    return x**2 + y**2

def f3(x, y):
    return np.sin(3 * x) * y
```

```python
# function 1
Z1 = f1(X, Y)
print(f'Shape Z1: {Z1.shape}')

# function 1
Z2 = f2(X, Y)
print(f'Shape Z2: {Z2.shape}')

# function 1
Z3 = f3(X, Y)
print(f'Shape Z3: {Z3.shape}')
```

```python
fig = plt.figure(figsize=(8, 6))
# 111 means 1 row, 1 col, 1 subplot
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1, cmap='viridis')
ax.set_title('f(x, y) = x * y')
plt.show()

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z2, cmap='viridis')
ax.set_title('f(x, y) = x * y')
plt.show()

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z3, cmap='viridis')
ax.set_title('f(x, y) = x * y')
plt.show()
```

```python
bars = np.loadtxt('./data/values_for_bars.csv', delimiter=',', skiprows=0)
hist = np.loadtxt('./data/values_for_hist.csv', delimiter=',', skiprows=0)

print(f'Shape bars: {bars.shape}')
print(f'Shape hist: {hist.shape}')
```

```python
unique_counts = Counter(bars)

unique_values = list(unique_counts.keys())
counts = list(unique_counts.values())
```

```python
plt.figure(figsize=(8, 4))
plt.bar(unique_values, counts, color='purple', edgecolor='k', alpha=0.75)
plt.xlabel('Unique Values')
plt.ylabel('Frequency')
plt.title('Bar plot for values_for_bars')
plt.grid(axis='y', linestyle='--', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
# divided into 30 equal parts: bins=30
plt.hist(hist, bins=30, color='purple', edgecolor='k', alpha=0.7)
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.title('Histogram of real numbers')
plt.grid(axis='y', linestyle='--', alpha=0.75)
plt.show()
```
